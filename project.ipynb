```python
import requests,json
from pylab import rcParams#出力画像のサイズ変更
from japanmap import *
%matplotlib inline
import matplotlib.pyplot as plt
rcParams['figure.figsize']=10,10
#-------------------------------------------------------------------------------------
API_KEY = "0a5357d6b04726d418aad4cca2202999"
api = "http://api.openweathermap.org/data/2.5/weather?q={city}&units=metric&lang=ja&appid={key}"
count=0
data={}
color=[]
pref_list=['北海道', '青森県', '岩手県','沖縄県']
pref_weather=["Hokkaidō","Aomori-ken","Iwate-ken","Okinawa"]
#--------------------------------------------------------------------------------------
def temp_color(a):
    if a >= 30:
        color='red'
        return color
    elif a>=25:
        color='orange'
        return color
    elif a>=20:
        color='yellow'
        return color
    elif a>15:
        color='lightgreen'
        return color
    elif a>10:
        color='green'
        return color
    elif a>=5:
        color='cyan'
        return color
    else:
        color='blue'
        return color

for city_name in pref_weather:
    url = api.format(city = city_name, key = API_KEY)
    response = requests.get(url)
    data = json.loads(response.text)
    temp=data["main"]["temp"]
    color.append(temp)
for l,w in zip(pref_list,color):#dataに辞書形式で代入
    col=temp_color(w)
    data[l]=col
    
plt.imshow(picture(data));

    
    
```
